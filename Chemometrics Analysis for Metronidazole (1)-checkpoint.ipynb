```python
#Extra wavenumber and spectra
Wavenumber=data['Wavenumber']
Spectra=data.drop('Wavenumber', axis=1)
```

```python
#1 Spectral preprocessing as per the order specified to ensure normality before Chemometric analysis
#1. baseline correction -polynomial fitting
#1 Spectral preprocessing as per the order specified to ensure normality before Chemometric analysis
#1. baseline correction -polynomial fitting
from scipy.signal import savgol_filter

# Assuming 'Spectra' is a DataFrame with columns representing different spectra,
# and 'Wavenumber' is a column representing the wavenumbers.

# Define a window size and polynomial order for Savitzky-Golay smoothing
window_size = 15  # Adjust as needed
poly_order = 2   # Adjust as needed

# Apply Savitzky-Golay smoothing to each spectrum
baseline_corrected_Spectra = Spectra.apply(lambda spectrum: savgol_filter(spectrum, window_size, poly_order), axis=0)
first_derivative_Spectra=baseline_corrected_Spectra.diff()
second_derivative_Spectra=first_derivative_Spectra.diff()
```

```python
#Apply Multiplicative Scatter Correction (MSC)

#The first step is to calculate the mean spectrum (average spectrum)
mean_spectrum=baseline_corrected_Spectra.mean(axis=1)

#Next we calculate the ratio of each spectrum to the mean spectrum

msc_corrected_Spectra=baseline_corrected_Spectra.divide(mean_spectrum,axis=0)
#Apply Standard Normal variate (SNV)
snv_corrected_Spectra = (baseline_corrected_Spectra - baseline_corrected_Spectra.mean(axis=1)[:, np.newaxis]) / baseline_corrected_Spectra.std(axis=1)[:, np.newaxis]
```

```python
#Plot original and preprocessed data 
plt.figure(figsize=(12,10))
plt.subplot(4,1,1)
plt.plot(Wavenumber, Spectra)
plt.title('Original Spectra')

plt.subplot(4,1,2)
plt.plot(Wavenumber, baseline_corrected_Spectra)
plt.title('Baseline Corrected Spectra')

plt.subplot(4,1,3)
plt.plot(Wavenumber,first_derivative_Spectra)
plt.title('First Derivative')

plt.subplot(4,1,4)
plt.plot(Wavenumber, second_derivative_Spectra)
plt.title('Second Derivative')

plt.tight_layout()
plt.show()
```

```python
#Apply Multiplicative Scatter Correction (MSC)

#The first step is to calculate the mean spectrum (average spectrum)
mean_spectrum=Spectra.mean(axis=1)

#Next we calculate the ratio of each spectrum to the mean spectrum

msc_corrected_Spectra=Spectra.divide(mean_spectrum,axis=0)

#We compare original and MCS -corrected spectra
plt.figure(figsize=(12,10))
plt.subplot(4,1,1)
plt.plot(Wavenumber, Spectra)
plt.title('Original Spectra')

plt.subplot(4,1,2)
plt.plot(Wavenumber, msc_corrected_Spectra)
plt.title('MSC Corrected Spectra')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'baseline_corrected_Spectra' is your DataFrame with baseline corrected spectral data

# Convert the DataFrame to a NumPy array
spectra_array = baseline_corrected_Spectra.to_numpy()

# Mean Spectrum
mean_spectrum = np.mean(spectra_array, axis=1)

# SNV Correction
snv_corrected_spectra = (spectra_array - mean_spectrum[:, np.newaxis]) / np.std(spectra_array, axis=1)[:, np.newaxis]

# Convert back to DataFrame if needed
snv_corrected_spectra_df = pd.DataFrame(snv_corrected_spectra, columns=baseline_corrected_Spectra.columns)

# Plot original and SNV-corrected spectra
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(baseline_corrected_Spectra.columns, spectra_array.T)
plt.title('Original Spectra')

plt.subplot(2, 1, 2)
plt.plot(baseline_corrected_Spectra.columns, snv_corrected_spectra_df.T)
plt.title('SNV-Corrected Spectra')

plt.tight_layout()
plt.show()
```
